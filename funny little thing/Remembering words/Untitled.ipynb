```python
import json
import MySQLdb
db = MySQLdb.connect(host = 'localhost',#本地数据库
                             user = 'root', #用户名
                             passwd = 'tw123123', #数据库密码
                             db = 'mydict', #数据库名
                             charset = 'utf8') 
output = open('Synonym.md','w',encoding='utf8')
```

```python
my_dic = open('data.json','r',encoding='utf8').read()
my_dic = json.loads(my_dic)
```

```python


def create_table(ch_def):
    sql = """call words_by_def('%s')"""%(ch_def)
    cursor = db.cursor()
    cursor.execute(sql)
    word_list = cursor.fetchall()
    word_list = [w[0] for w in word_list]

    res = '|word|meaning|word|meaning|\n|---|---|---|---|\n'

    for i in range(len(word_list)):
        w = word_list[i]
        print(w)
        w_info = my_dic[w]
        meaning = w_info['chinese'].strip().replace('\n','<br/>')
        if i%2 != 0:
            res = res + w + '|' + meaning + '|\n'
        else:
            res = res + '|' + w + '|' + meaning + '|'
    if i % 2 == 0:
        res = res + '| | |\n'

    res = '**' + ch_def + ':**\n' + res + '\n\n'
    return res


def write_md(ch_def):
    output = open('Synonym.md','w',encoding='utf8')
    output.write('# Synonyms\n\n')
    for d in ch_def:
        res = create_table(d)
        output.write(res)
    output.close()
    
```

```python
ch_def = ['责骂|斥责|训斥|指责','嘲笑','粗鲁','和蔼','柔软',
          '凶残|恶毒|残暴','贪婪','鲁莽|草率','固执','陈腐|平庸|陈词滥调',
         '乞求|恳求|请求','模糊','难懂|深奥','预感','过时|废弃的',
          '傲慢','混乱','镇定|镇静','诽谤','新生的']
write_md(ch_def)
```
